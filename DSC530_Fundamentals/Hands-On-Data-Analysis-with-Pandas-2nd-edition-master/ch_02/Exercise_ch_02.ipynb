```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/parsed.csv')

df.head()
```

1. Find the 95th percentile of earthquake magnitude in Japan using 'mb' magnitude type

```python
# filter for 'mb' magnitude type and location containing 'Japan'
# the .str.contains('Japan', case=Talse) ensure the filter is case insensitive
# na=False is included to handle any potential missing 'place' values

japan_mb_magnitudes = df[
    (df['magType'] == 'mb') &
     (df['place'].str.contains('Japan', case=False, na=False))
]['mag']

# Take the 95th percentile (0.95 quantile) of the filtered magnitudes
percentile_95 = japan_mb_magnitudes.quantile(0.95)

print(f"The 95th percentile of 'mb' magnitude in Japan is: {percentile_95}")
```

2. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.

```python
# filter the DataFrame for earthquakes in Indonesia
# the. str.contains() method checks if 'Indonesia" is in the place string
# Case=False insensitive for case

indonesia_df = df[
    df['parsed_place'].str.contains('Indonesia', case=False, na=False)
    ]

# Finding the number of rows with shape
total_indonesia_events = indonesia_df.shape[0]

print(f"Indonesia had {total_indonesia_events} earthquakes")

# Finding the tsunami amount in Indonesia 'tsunami '==1 using the sum of all events. No even is 0
tsunami_events = indonesia_df['tsunami'].sum()

print(f"Out of {total_indonesia_events} earthquakes {tsunami_events} included tsunami.")

# Calculating percentage
percentage = (tsunami_events / total_indonesia_events) * 100


print(f"Percentage of earthquakes with tsunami in Indonesia: {round(percentage,2)}")
```

4. Add a column indicating whether the earthquake happened in a country or US state that is on the Ring of Fire. Use Alaska, Antarctica (look for Antarctic), Bolivia, California, Canada, Chile, Costa Rica, Ecuador, Fiji, Guatemala, Indonesia, Japan, Kermadec Islands, Mexico (be careful not to select New Mexico), New Zealand, Peru, Philippines, Russia, Taiwan, Tonga, and Washington.

```python
# Define list of Ring of Fire locations
# Antarctica is Antarctic and New Mexico is excluded '(?<!New )Mexico'

ring_of_fire_locations = [
   'Alaska', 'Antarctic', 'Bolivia', 'California', 'Canada', 'Chile', 'Costa Rica', 'Ecuador', 'Fiji', 'Guatemala', 'Indonesia', 'Japan',
    'Kermadec Islands', '(?<!New )Mexico', 'New Zealand', 'Peru', 'Philippines', 'Russia', 'Taiwan', 'Tonga', 'Washington'
]

# Create the regex pattern joining the list with the OR operator (|)
regex_pattern = '|'.join(ring_of_fire_locations)

# Create new df with ring of fire
# Using regex=True and na=False to handle missing values and ensure regex matching
df_with_ring_of_fire = df.assign(
    in_ring_of_fire=df['parsed_place'].str.contains(
        regex_pattern,
        case=False,  # case-insensitive matching
        na=False,
        regex=True  # use regex matching
    )
)

print(df_with_ring_of_fire[['parsed_place', 'in_ring_of_fire']].head())

# Creating sample
sample_df = df_with_ring_of_fire[['title', 'parsed_place', 'in_ring_of_fire']].sample(5)
print(sample_df)
```

5. Calculate the number of earthquakes in the Ring of Fire locations and the number outside of them.

```python
# print out the total count of Ring of Fire earthquakes
ring_of_fire_count = df_with_ring_of_fire['in_ring_of_fire'].sum()
not_in_ring_of_fire = (~df_with_ring_of_fire['in_ring_of_fire']).sum()
total_records = len(df_with_ring_of_fire)
print(f"\nTotal Earthquake locations: {total_records}")
print(f"\nTotal Earthquakes in Ring of Fire locations: {ring_of_fire_count}")
print(f"\nTotal Earthquakes not in Ring of Fire locations: {not_in_ring_of_fire}")
```
