# Chapter 3, Exercise 1
We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks, but we were given each as a separate CSV file (obtained using the stock_analysis package we will build in Chapter 7, Financial Analysis – Bitcoin and the Stock Market). Combine them into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises:

Read in the aapl.csv, amzn.csv, fb.csv, goog.csv, and nflx.csv files.
Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example); this is how you look up a stock. In this case, the filenames happen to be the ticker symbols.
Append them together into a single dataframe.
Save the result in a CSV file called faang.csv

```python
# Importing all the data frames
# Assigning ticker to each data frame
import pandas as pd
aapl_df= pd.read_csv('exercises/aapl.csv').assign(ticker='AAPL')
amzn_df= pd.read_csv('exercises/amzn.csv').assign(ticker='AMZN')
fb_df= pd.read_csv('exercises/fb.csv').assign(ticker='FB')
goog_df= pd.read_csv('exercises/goog.csv').assign(ticker='GOOG')
nflx_df= pd.read_csv('exercises/nflx.csv').assign(ticker='NFLX')

# Combining the data frames into one list
all_df = [aapl_df, amzn_df, fb_df, goog_df, nflx_df]

# Using pd.concat() to append the rows of all data frames into one data frame
# Ignoring the index in each data frame and resetting it into the final FAANG data frame
faang_df = pd.concat(all_df, ignore_index=True)

# saving the result as CSV file called faang.csv
faang_df.to_csv('exercises/faang.csv', index=False)


# taking a sample of the new data frame.
faang_df.sample(n=5, ignore_index=True)

```

Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.

```python
# setting the format due to dollar amounts were showing in scientific notation
pd.set_option('display.float_format','{:,.2f}'.format)

# Use melt() to make it completely long format
melted_df = faang_df.melt(
    id_vars = ['date', 'ticker'], value_vars=['open', 'high', 'low', 'close', 'volume'],
    value_name= 'dollars', var_name='measure'
)

# Sorting the df by data, ticker, and measure
sorted_melted_df = melted_df.sort_values(
    ['date','ticker', 'measure']
).reset_index(drop=True)

sorted_melted_df.head(n=10)
```

# Chapter 3, Exercise 5
Suppose we found out that on July 26, 2018 there was a glitch in how the data was recorded. How should we handle this? Note that there is no coding required for this exercise.

To initiate an investigation of an anomaly, it is essential to confirm that a glitch exists and define its scope. We need to find out:
Scope of Impact: Was the anomaly present only in one ticker, or did it affect all of them?
Temporal Range: Did the glitch infect only one day, or did it also affect previous or following days?
Origin: Did the glitch originate in one of the original source data files, or did it appear during the file combination process?
If the glitch originated from one of the source data files, we have several options for correction:
Option 
1: Exclusion. We could exclude or filter out data from the affected day to avoid bias. However, losing data may negatively impact short-term analysis.
Option 
2: Imputation (Fill/Interpolation). We could use forward or backward fill, replacing the glitched value with the closing price from the previous day (e.g., July 25th) or the next day (e.g., July 27th). Interpolation is another option, which replaces the missing value by calculating the average of the surrounding values. While this allows the data to run smoothly, the replaced data will be artificial and may obscure a real-world market event that appears as a glitch.
Option 
3: External Verification. If time allows, we could seek the correct data from another verified source to replace the flawed data entirely.
In any case, it is essential to be transparent and thoroughly document the changes made to the original data. This notation should clarify the date the glitch occurred, its nature, and the specific method used to address it



# Chapter 3, Exercise 6
The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:

Read in the covid19_cases.csv file.
Create a date column using the data in the dateRep column and the pd.to_datetime() function.
Set the date column as the index and sort the index.
Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively. Hint: the replace() method can be run on the dataframe as a whole.
Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.
Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). Be sure to fill in NaN values with 0.

```python
# reading in covid19_cases.csv
covid_data = pd.read_csv('exercises/covid19_cases.csv')

covid_data.head(5)
```

```python
# Creating a date column using dateRep column and pd.to_datetime() function
covid_data.loc[:, 'date'] = pd.to_datetime(
    covid_data['dateRep'], 
    dayfirst=True,
)
```

```python
# Set the 'date' column as the index and sort
covid_data = covid_data.set_index('date').sort_index()

# checking data
covid_data.head(5)
```

```python
#Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively. Hint: the replace() method can be run on the dataframe as a whole.
replacements = {
    'United_States_of_America': 'USA',
    'United_Kingdom': 'UK'
}
covid_data = covid_data.replace(replacements, regex=False)


```

```python
# Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to 
# Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.
filtered_covid_data = covid_data[
covid_data['countriesAndTerritories'].isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India',
                                           'Italy', 'Mexico', 'Peru', 'Russia', 'Spain',
                                           'Turkey', 'UK', 'USA'])
]
filtered_covid_data[filtered_covid_data['countriesAndTerritories'] == 'UK'].head(5)
```

```python
# Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). 
cases_by_country = filtered_covid_data.reset_index().pivot_table(
    index='date',
    columns='countriesAndTerritories',
    values='cases'
)

# Be sure to fill in NaN values with 0.
cases_by_country = cases_by_country.fillna(0)

cases_by_country.head()
```
