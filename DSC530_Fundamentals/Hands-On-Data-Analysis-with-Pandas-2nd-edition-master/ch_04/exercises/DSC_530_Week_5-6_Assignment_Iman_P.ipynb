#Exercise 1.

With the earthquakes.csv file, select all the earthquakes in Japan with a magnitude of 4.9 or greater using the mb magnitude type.

```python
import numpy as np
import pandas as pd

earthquakes = pd.read_csv('earthquakes.csv')

earthquakes.head() #checked for headings in data
```

```python
#Choosing earthquakes where the place equals 'Japan,' and the magnitude is 4.9 or bigger, and where mag type is 'mb'
japan_mb_earthquakes = earthquakes[(earthquakes.parsed_place == 'Japan') & (earthquakes.mag >= 4.9) & (earthquakes.magType =='mb')]
japan_mb_earthquakes
```

Exercise 2.

Create bins for each full number of earthquake magnitude (for instance, the first bin is (0,1], the second is (1,2], and so on) with the ml magnitude type and count how many are in each bin.

```python
#selecting earthquakes with magType ml
ml_earthquakes = earthquakes[earthquakes.magType == 'ml']

#finding the maximum mag value using np.ceil(), assigning it as int
mag_max = int(np.ceil(ml_earthquakes.mag.max()))

#creating a list for bins where the largest value is included
bins = list(range(0, mag_max +1))

#using pd.cut to assign the values in bins starting from (0,1] (including 1)
binned_ml_counts = pd.cut(ml_earthquakes.mag, bins = bins).value_counts().sort_index()

#printing out the result
print(binned_ml_counts)
```

Exercise 4.

Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magnitude type along the columns.

```python
#using crosstab() using tsunami as rows and magType as columns
#aggfunc='max' finds the largest magnitude of earthquakes
mag_max_crosstab = pd.crosstab(index=earthquakes.tsunami, columns=earthquakes.magType, values=earthquakes.mag, aggfunc='max')
mag_max_crosstab
```

Create a pivot table of the FAANG data that compares the stocks. Put the ticker in the rows and show the averages of the OHLC and volume traded data.

Exercise 7.

Calculate the z-scores for each numeric column of Amazon’s data (ticker is AMZN) in Q4 2018 using apply().



```python
#importing zscore from scipy.stats
#z-score = (value-mean)/std
from scipy.stats import zscore

#filtering data for Amazon and Q4 2018
amzn_q4 = faang_data[(faang_data.ticker == 'AMZN') & (faang_data.date >= '2018-10-01') & (faang_data.date <= '2018-12-31')].copy()

amzn_q4.head() #checked the data header
```

The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data through September 18, 2020. Complete the following tasks to practice the skills you've learned up to this point in the book:



Prepare the data

Read in the data in the covid19_cases.csv file.

Create a date column by parsing the dateRep column into a datetime.

Set the date column as the index.

Use the replace() method to update all occurrences of United_States_of_America and United_Kingdom to USA and UK, respectively.
Sort the index.

For the five countries with the most cases (cumulative), find the day with the largest number of cases.

Find the 7-day average change in COVID-19 cases for the last week in the data for the five countries with the most cases.

Find the first date that each country other than China had cases.

Rank the countries by cumulative cases using percentiles.

```python
# Reading in data
Covid_df = pd.read_csv('covid19_cases.csv')
Covid_df.head()
```

```python
# Parsing dateRep into datetime and setting it as index
Covid_df['date'] = pd.to_datetime(Covid_df['dateRep'], format = '%d/%m/%Y')
Covid_df.set_index('date', inplace=True)

#Using replace() to update country names
Covid_df['countriesAndTerritories'] = Covid_df['countriesAndTerritories'].replace({
    'United_States_of_America': 'USA',
    'United_Kingdom': 'UK'})

# Sorting the index
Covid_df.sort_index(inplace=True)
Covid_df.head()
```

```python
#Finding the five countries with the most cases (cumulative)
top_5_countries  = Covid_df.groupby('countriesAndTerritories')['cases'].sum().nlargest(5).index.tolist()
top_5_countries
```

```python
# finding the day with the largest number of cases for these 5 countries.
max_cases_day = Covid_df[Covid_df['countriesAndTerritories'].isin(top_5_countries)].groupby('countriesAndTerritories')['cases'].idxmax()

print(max_cases_day)
```

```python
#Find the 7-day average change in COVID-19 cases for the last week in the data for the five countries with the most cases.
#Finding the last 7 days
last_date = Covid_df.index.max()
#Finding last week data
last_week = Covid_df[Covid_df['countriesAndTerritories'].isin(top_5_countries)].loc[last_date - pd.Timedelta(days=6):last_date]
avg_change = last_week.groupby('countriesAndTerritories')['cases'].mean()
avg_change.round(2)
```

```python
#Find the first date that each country other than China had cases.
first_dates = Covid_df[(Covid_df['countriesAndTerritories'] != 'China') & (Covid_df['cases'] > 0)].groupby('countriesAndTerritories').apply(lambda x: x.index.min())
print(first_dates)
```

```python
#Rank the countries by cumulative cases using percentiles.
cumul_percentile = Covid_df.groupby('countriesAndTerritories')['cases'].sum()
rank_percentile = cumul_percentile.rank(pct=True).sort_values(ascending=False)
print(rank_percentile.round(4))
```
