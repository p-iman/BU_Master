Create a list, iterate over the list and sort your results, generate random numbers, add to the list, and then print your results.

```python
import random
#Creating a list using random.sample to have all unique values.
number_list_1 = [random.randint(0,300) for x in range (0,20)]
print(number_list_1)
```

```python
#Iterating over the list and sorting the result
for x in sorted(number_list_1):
    print(x)
```

```python
#Creating a second list to add to the list and creating a new list.
number_list_2 = [random.randint(300,400) for x in range (0,10)]
number_list_3 = number_list_1 + number_list_2
for x in sorted(number_list_3):
    print(x)
```
