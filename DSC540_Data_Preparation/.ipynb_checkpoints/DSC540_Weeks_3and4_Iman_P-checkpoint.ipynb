```python
#9. Plot log10(crime) vs. price
plt.scatter(np.log10(df_small['CRIM']), df_small['PRICE'], c = 'purple')
plt.title(" Crime rate vs. Price" , fontsize = 16)
plt.xlabel("Crime Rate (log10)", fontsize = 12)
plt.ylabel("Median Home Price", fontsize = 12)
plt.show()
```

```python
#10. Calculating a few statistics
#Mean room per dwelling
rooms = df_small['RM'].mean()
print(f"Houses have average of {rooms: .2f} rooms per dwelling.")
```

```python
# Medina age
median_age = df_small['AGE'].median()
print(f"Medina age is {median_age}.")
```

```python
# Mean distance 
average_dis = df_small['DIS'].mean()
print(f"Average distance is {average_dis: .2f}.")
```

```python
# Percentage of with low prices where price <$20,000
low_price = df_small['PRICE']<20
print(low_price)
```

```python
#4. Add the name of Income for the response variable to the dataset
names.append('Income')
names
```

```python
df = pd.read_csv(income_data_path,names=names)
df.head(5)
```

```python
#5. Finding the missing values
# seeking the count of missing values (NaN) for each column
print(df.isnull().sum())
```

```python
# The data frame does not have missing values.

#6. Create a DataFrame with only age, education, and occupation by using subsetting
df_subset = df[['age','education', 'occupation']]
df_subset.head()
```

```python
#7. Plot a histogram of age with bin size of 20
df_subset['age'].hist(bins = 20, edgecolor = 'black')
plt.title("Age Data Histogram", fontsize = 10)
plt.show
```

```python
#8. Creating a function to strip the whitespace characters. This can be used only for the columns with strings.
def strip_whitespace(c):
    return c.strip()
```

```python
# 9. Use the apply method to apply this function to all the columns with string values. 
# Create a new column
# Copy the values from this new column to the old column.
# Drop the new column
# Column: education
df_subset['education_stripped'] = df_subset['education'].apply(strip_whitespace)
df_subset['education'] = df_subset['education_stripped']
df_subset.drop(labels= ['education_stripped'], axis=1, inplace=True)
```

```python
# Column: occupation
df_subset['occupation_stripped'] = df_subset['occupation'].apply(strip_whitespace)
df_subset['occupation'] = df_subset['occupation_stripped']
df_subset.drop(labels= ['occupation_stripped'], axis=1, inplace=True)
```

```python
#10. Find the number of people who are aged between 30 and 50
df_filtered = df_subset[(df_subset['age']>=30) & (df_subset['age']<= 50)]
number = df_filtered.shape[0]
print(f"There are {number} people between 30 and 50 years of age.")
```

```python
#11. Group the records by age and education to determine how the mean age is distributed. 
by_age_education = df_subset.groupby('education')['age'].mean()
print(by_age_education.sort_values().round(2))
```

```python
#12. Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on average 
#and which profession has its largest share of the workforce above the 75th percentile.
occup_summary = df.groupby('occupation')['age'].describe()
print(occup_summary.round(2))
```

```python
#Oldest workers
oldest_profession =occup_summary.sort_values(by='mean', ascending=False)
print(oldest_profession.round(2))
```

```python
#Highest 75%
oldest_75th =occup_summary.sort_values(by='mean', ascending=False)
print(oldest_75th.round(2))
```

```python
#13. Use subset and groupby ot find the outliers
occupation_statistics= df_subset.groupby(
    'occupation').describe()['age']
print(occupation_statistics)
```

```python
#14. Plot the outlier values on a bar chart. 
plt.figure(figsize=(15,8))
plt.barh(y=occupation_stats.index,
         width=occupation_stats['count'])
plt.yticks(fontsize=13)
plt.show()
```

```python
# 15. Merging two Data Frames using common keys.
#creating df_1 including columns age, workclass and occupation
df_first = df[['age', 'workclass', 'occupation']].sample(5,random_state=101)
df_first.head()
```

```python
# creating the second data frame to add education
df_second = df[['education',
           'occupation']].sample(5,random_state=101)
df_second.head()
```

```python
# Merging the data frames together.
df_third = pd.merge(df_first,df_second, on='occupation', how='inner').drop_duplicates()
df_third
```

# 3. Create a series and practice basic arithmetic steps


a. Series 1 = 7.3, -2.5, 3.4, 1.5

          i. Index = ‘a’, ‘c’, ‘d’, ‘e’

b. Series 2 = -2.1, 3.6, -1.5, 4, 3.1

          i. Index = ‘a’, ‘c’, ‘e’, ‘f’, ‘g’

c. Add Series 1 and Series 2 together and print the results

d. Subtract Series 1 from Series 2 and print the results

```python
# a. 
series1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
#b
series2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
# c. 
result = series1 + series2
print("Result of adding the series1 and series2 together")
print(result)
```

```python
#d
result2 = series2 - series1
print("Result of subtracting the series1 from series2 ")
print(result2)
```
