```python
# 1. Creating 100 random number list. Random was imported in the first task in this notebook.
random_numbers = [random.randint(0, 100) for x in range (0,100)]
random_numbers
```

```python
# 2. Creating a list where numbers are divisible by 3
list_divisible_by_3 = [n for n in random_numbers if n % 3 == 0]
list_divisible_by_3
```

```python
# 3. Calculating lengths of the lists and the difference in their lengths
length_of_random_numbers = len(random_numbers)
length_of_list_divisible_by_3 = len(list_divisible_by_3)
difference = length_of_random_numbers - length_of_list_divisible_by_3

difference
```

```python
# 4. Performing the last 3 tasks 10 times by using a loop.
number_of_experiments = 10
difference_list = []
for i in range(0, number_of_experiments):
    random_numbers = [random.randint(0, 100) for x in range (0, 100)]
    list_divisible_by_3 = [n for n in random_numbers if n % 3 == 0]

    length_of_random_numbers = len(random_numbers)
    length_of_list_divisible_by_3 = len(list_divisible_by_3)
    difference = length_of_random_numbers - length_of_list_divisible_by_3
    difference_list.append(difference)

difference_list
```

```python
# Calculating the average difference

avg_diff = sum(difference_list) / float(len(difference_list))
avg_diff
```

b. The Data Wrangling Workshop: Activity 1.02 page 43

```python
# 1. Creating the multiline text variable using the provided quote
multiline_text= """It is a truth universally acknowledged, that a single man in possession of a good fortune, must be in want of a wife.

However little known the feelings or views of such a man may be on his first entering a neighbourhood, this truth is so well fixed in the minds of the surrounding families, that he is considered the rightful property of some one or other of their daughters.

"My dear Mr. Bennet," said his lady to him one day, "have you heard that Netherfield Park is let at last?"

Mr. Bennet replied that he had not.

"But it is," returned she; "for Mrs. Long has just been here, and she told me all about it."

Mr. Bennet made no answer.

"Do you not want to know who has taken it?" cried his wife impatiently.

"You want to tell me, and I have no objection to hearing it."

This was invitation enough.

"Why, my dear, you must know, Mrs. Long says that Netherfield is taken by a young man of large fortune from the north of England; that he came down on Monday in a chaise and four to see the place, and was so much delighted with it, that he agreed with Mr. Morris immediately; that he is to take possession before Michaelmas, and some of his servants are to be in the house by the end of next week."

"What is his name?"""
```

```python
# 2. Finding the type and the length of the text
type(multiline_text)
```

```python
len(multiline_text)
```

```python
#3. Removing the lines and symbols using the replace method.
multiline_text = multiline_text.replace('\n', "")
multiline_text
```

```python
cleaned_text = ""
for char in multiline_text:
    if char == " ":
        cleaned_text += char
    elif char.isalnum(): 
        cleaned_text += char
    else:
        cleaned_text += " "
cleaned_text
```

```python
# 4. Finding all the words using the split method
cleaned_text_words = cleaned_text.split()
cleaned_text_words
```

```python
#5. Creating a list of unique words
unique_words_as_dict = dict.fromkeys(cleaned_text_words)
len(list(unique_words_as_dict.keys()))
```

```python
#6. Counting how many times each word appears in the cleaned text.
for word in cleaned_text_words:
    if unique_words_as_dict[word] is None:
        unique_words_as_dict[word] = 1
    else:
        unique_words_as_dict[word] += 1

unique_words_as_dict
```

```python
#7. Finding the 25 most common words
top_25_words = sorted(unique_words_as_dict.items(), key=lambda key_val_tupple: key_val_tupple[1], reverse=True)
top_25_words[:25]
```

```python
#6. Combining code into one block and creating a number by reading individual digits from the stack
import math
def convert_to_number(number_stack):
    final_number = 0
    for digit in number_stack:
        final_number = final_number * 10 + digit
    return float(final_number)

for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile (lambda x: x <+ 0, number_tuple))
    print(convert_to_number(number_stack))
```

d. The Data Wrangling Workshop: Activity 2.02 page 81

```python
#1. Importing zip_longest from itertools, and creating a function to zip header, line, and fillvalue = None.
from itertools import zip_longest

def return_dict_from_csv_line(header, line):
    # Pairing up items from the header and the line
    zipped_line = zip_longest(header, line, fillvalue=None)
    # generating the directory for every tupple in the zipped list
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict  
```

```python
# Reading the file line by line
# creating filepath
file_path = r'C:\Users\paula.iman\Documents\BU_Master\DSC540_Data_Preparation\The-Data-Wrangling-Workshop-master\Chapter02\datasets\sales_record.csv'
with open(file_path, "r") as fd:
    first_line = fd.readline()
    # remove the hidden new line
    header = first_line.replace("\n", "").split(",")
    # keep count  counter i starting from 0
    for i, line in enumerate(fd):
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i >10:
            break # adding safety break
```
